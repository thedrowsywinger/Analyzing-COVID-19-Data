Some of these papers have a list of abstract with a length as high as 29

```python
number_of_abstracts = 0
for i in list_of_files:
    with open(i, "rb") as f:
        s = json.load(f)
    if len(s['abstract']) > number_of_abstracts:
        number_of_abstracts = len(s['abstract'])
        
print("Higest number of abstracts: ", number_of_abstracts)
```

I have also decided to take one abstract per paper , and if a paper doesn't have an abstract , then i'm going to discard it completely

right now i plan to just take the first abstract text , i'll check later for missing ones

```python
list_of_abstracts = []
list_of_body = []
counter = 0
for i in tqdm(list_of_files):
    with open(i, "rb") as f:
        s = json.load(f)
    list_of_abstracts.append(s['abstract'][0]['text'])
    counter += 1
    entire_body = ""
    for j in s['body_text']:
          entire_body += j['text'] + "\n\n"  
```

So there may be cases where the abstract list is actually empty

```python
list_of_abstracts = []
list_of_body = []
counter = 0
for i in tqdm(list_of_files):
    with open(i, "rb") as f:
        s = json.load(f)
    if len(s['abstract']) != 0:
        list_of_abstracts.append(s['abstract'][0]['text'])
        entire_body = ""
        for j in s['body_text']:
            entire_body += j['text'] + "\n\n"  
        list_of_body.append(entire_body)
```
